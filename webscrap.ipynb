```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
url = "https://m.imdb.com/chart/tvmeter/?ref_=hm_nv_menu/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/122.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
print(response.status_code)
print(response.text[:500])  # preview first 500 chars
```

```python
soup = BeautifulSoup(response.text, 'html.parser') 
```

```python
movies = soup.find("ul",class_ = "ipc-metadata-list ipc-metadata-list--dividers-none sc-d24d5d37-0 hDHQeM alt_item_colors compact-list-view ipc-metadata-list--base")
```

```python
headers = ["id", "title", "release_year", "rating"]
with open("imdb_titles.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(headers)

    id_show = 0

    for movie in movies:
        id_show += 1

        title = movie.find("a", class_="ipc-title-link-wrapper").text.strip()
        release_year = movie.find( "span",class_="sc-b4f120f6-7 hoOxkw cli-title-metadata-item").text.strip()
        rating_tag = movie.find("span", class_="ipc-rating-star--rating")

        if rating_tag is not None:
            rating = rating_tag.text.strip()
        else:
            rating = "N/A"

        print(id_show, title, release_year, rating)

        writer.writerow([
                id_show,
                title,
                release_year,
                rating
            ])
```
